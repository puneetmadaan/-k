````python
import pandas as pd
def create_comparison_df(map_25, map_31, case_key):
    # Extract raw dictionaries from DSPy Prediction objects
    data_25 = map_25[case_key]._store
    data_31 = map_31[case_key]._store

    # Create the DataFrame
    df = pd.DataFrame([data_25, data_31], index=[f'Gemini 2.5 (Clerk) : {case_key}', f'Gemini 3.1 (Guardian) : {case_key}']).T

    # Add a "Delta / Difference" column for the numeric biometrics
    # This highlights the "Functional Deafness" gap
    metrics = ['pitch_jitter', 'noun_to_verb_ratio', 'placeholder_density']
    df['Audit Delta'] = ""

    for m in metrics:
        if m in df.index:
            diff = abs(df.loc[m, f'Gemini 2.5 (Clerk) : {case_key}'] - df.loc[m, f'Gemini 3.1 (Guardian) : {case_key}'])
            df.at[m, 'Audit Delta'] = f"Δ {diff:.3f}"

    return df
````
